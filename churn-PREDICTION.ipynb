```python
import pandas as pd
data = pd.read_csv('ass.csv')
data.head(10)
```

```python
data.info()
```

```python
# Analysing data
from sklearn.preprocessing  import LabelEncoder
for i in list(data.columns):
  print(i, end=": ")
  print(set(data[i].values.tolist()))
  print()

# converting categorical data into numerical data
for row in ['canceled_service','senior_citizen','spouse_partner', 'dependents', 'cellular_service', 'internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies', 'contract', 'paperless_bill', 'payment_method']:
  label = LabelEncoder().fit_transform(data[row])
  data.drop(row, axis=1, inplace=True)
  data[row] = label
```

```python
# After converting into numerical data
for i in list(data.columns):
  print(i, end=": ")
  print(set(data[i].values.tolist()))
  print()
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(data, test_size = 0.2)
print(train.shape)
print(test.shape)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
data.corr()['canceled_service'].sort_values(ascending = False).plot(kind='bar')
```

```python
from sklearn.metrics.cluster import mutual_info_score
def compute_mutual_information(categorical_series):
  return mutual_info_score(categorical_series, data.canceled_service)

categ_var = data.select_dtypes(exclude = 'float64').drop('canceled_service', axis=1)
feature_imp = categ_var.apply(compute_mutual_information).sort_values(ascending =False)
feature_imp.plot.barh()
```

```python
list1 = ['internet_service','payment_method','tech_support','online_security','contract','avg_intl_mins','months_with_company','avg_call_mins']

train_x = train[list1]
train_y = train.canceled_service

test_x = test[list1]
test_y = test.canceled_service
train_x.head()
```

```python
train_y.head()
```

```python
# SVM
from sklearn import svm,metrics
model = svm.SVC()
model.fit(train_x,train_y)
predictionSVM = model.predict(test_x)
print("Accuracy: ",metrics.accuracy_score(predictionSVM,test_y))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(test_y,predictionSVM)
print(cm)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000) # max_iter avoids ConvergenceWarning
model.fit(train_x, train_y)
predictionLR = model.predict(test_x)


print("Accuracy: ", metrics.accuracy_score(predictionLR,test_y))
```

```python
cm=confusion_matrix(test_y,predictionLR)
print(cm)
```

```python
# K-Nearest neighbours
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3) # examines 3 neighbours
model.fit(train_x, train_y)
predictionKNN = model.predict(test_x)
print("Accuracy: ", metrics.accuracy_score(predictionKNN, test_y))
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(test_y,predictionKNN)
print(cm)
```
